6) 반복문을 이용하여 100이하의 소수(prime number)를 리스트에 담아 봅시다.
* 소수(prime number) : 약수가 오직 1과 자기자신 만 존재하는 수. 

```python
prime_list = []
for i in range(1,101):
    for j in range(2, i+1):
        if (j == i):
            prime_list.append(i)
        elif (i % j == 0):
            break
          
prime_list
```
